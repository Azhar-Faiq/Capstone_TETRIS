```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.interpolate import make_interp_spline
import numpy as np
from textwrap import wrap
```

```python
data_tahunan = pd.read_csv('data_tahunan.csv')
data_tahunan
```

```python
plt.plot(data_tahunan['tahun'],data_tahunan['nilai_ekspor'],color='g',label='Nilai Ekspor')
plt.plot(data_tahunan['tahun'],data_tahunan['nilai_impor'],color='red',label='Nilai Impor')
plt.legend()
plt.title("nilai ekspor impor")
plt.show
```
